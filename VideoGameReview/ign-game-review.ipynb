## 1.  Importing Data with Pandas

```python
import pandas as pd

# Read ign.csv into a DataFrame, and assign the result to reviews
reviews = pd.read_csv("ign.csv")

# prints the first N rows of a DataFrame
reviews.head()
```

```python
reviews.tail()
```

```python
# How many rows and columns in the data file
reviews.shape
```

## 2. Indexing DataFrames with Pandas:  

**Method 1: iloc\[r1:r2, c1:c2\]** 

```python
# I want to retrieve the first 5 rows with all columns
# Previously I used reviews.head()

reviews.iloc[0:5, :]
```

```python
# The first column is useless, so I will remove it

reviews = reviews.iloc[:, 1:]
reviews.head()
```

**Method 2: loc\[r1:r2, c1:c2\]**    
This allows us to index using labels instead of positions.

```python
reviews.loc[0:5, :]
```

```python
# I can retrieve several columns by their labels.

reviews.loc[25:30, ["score_phrase", "title", "score", "genre"]]
```

```python
# If I want to retrieve a whole column, I can do
# reviews.loc[:, "title"] or

reviews["title"].head()
```

## 3. How to filter data using Boolean operators:  

```python
# I want to see which of the rows have a score value higher than 7

score_filtered = reviews["score"] > 7
score_filtered.head()
```

```python
score_filtered.loc[25:35]
```

```python
# I want to only select rows in reviews where score is greater than 7

reviews_filtered = reviews[score_filtered]
reviews_filtered.head()
```

```python
reviews_filtered.loc[25:35]
```

```python
# I want to find games released for the iPad that have a score of more than 7

iPad_filtered = (reviews["platform"] == "iPad") & (reviews["score"] > 7)
reviews_iPad_filtered = reviews[iPad_filtered]
reviews_iPad_filtered.head()
```

```python
Macintosh_filtered = (reviews["platform"] == "Macintosh") & (reviews["score_phrase"] == "Amazing")
reviews_Macintosh_filtered = reviews[Macintosh_filtered]
reviews_Macintosh_filtered.head()
```

## 4. How to plot using Pandas:

```python
%matplotlib inline

iPad_filtered_1 = reviews["platform"] == "iPad"
reviews_iPad = reviews[iPad_filtered_1]

# plot a histogram for iPad's scores

reviews_iPad["score"].plot(kind="hist")
```

```python
# plot a histogram for Macintosh's scores

reviews[reviews["platform"] == "Macintosh"]["score"].plot(kind="hist", bins=20)
```
